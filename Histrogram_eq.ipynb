```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('hh.jpg')
plt.imshow(img)
plt.title("Original image")
plt.figure()
plt.hist(img.ravel(),256,[0,256]); 
plt.show()
```

```python
img = cv.imread('hh.jpg',0)
I_eq_cv2 = cv2.equalizeHist(img)

plt.figure()
plt.imshow(I_eq_cv2, cmap="gray", vmin=0, vmax=255)
plt.title(" result image")
plt.show()
plt.hist(I_eq_cv2.ravel(),256,[0,256]); 
plt.show()
```
